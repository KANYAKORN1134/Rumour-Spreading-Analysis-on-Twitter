```python
# @copyright: Thammachet Asavathongkul 2022 All rights reserved.

def myprint(d, output_structure,current_neasted_dict_path=[]):
    for k, v in d.items():
        current_neasted_dict_path.append(k)
        if not isinstance(v, dict):
            current_structure = None
            for neasted_index, key  in enumerate(current_neasted_dict_path):
                if neasted_index == 0:
                    if key not in output_structure:
                        output_structure[key] = {}
                    current_structure = output_structure[key]
                else:
                    if key not in current_structure:
                        current_structure[key] = {}
                    if len(current_neasted_dict_path) - 1 == neasted_index:
                        tweet_id = key
                        reaction_json = read_reaction(current_neasted_dict_path[0], tweet_id)
                        current_structure[key]['reaction'] = reaction_json
                    else:
                        current_structure = current_structure[key]
                        reaction_json = read_reaction(current_neasted_dict_path[0], key)
                        current_structure['reaction'] = reaction_json
    
        else:
            myprint(v,output_structure,current_neasted_dict_path)
        current_neasted_dict_path.pop()      

# @copyright: Thammachet Asavathongkul 2022 All rights reserved.
  
```

```python
# @copyright: Thammachet Asavathongkul 2022 All rights reserved.

def read_reaction(tweet_folder,reaction_id):
    # Read reaction json file
    with open(os.path.join(root_folder,tweet_folder,reaction_foldername,reaction_id+'.json'),'r') as f:
        reaction_json = json.load(f)
    return reaction_json

```
